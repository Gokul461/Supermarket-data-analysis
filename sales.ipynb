**Importing numpy and pandas libraries**

import numpy as np
import pandas as pd

**Loading sales data**

```python
d = pd.read_csv("Sales (1).csv")
d
```

```python
d1 = pd.DataFrame(d) # unique gender in data
print(d1['Gender'].unique())
```

**Checking for null values**

```python
nan_check = d1.isnull()
print(nan_check)
```

```python
d1.dropna(inplace=True)
print(d1)# dropping the null values
```

```python
d1.fillna(0, inplace=True)
print(d1)
```

```python
d1.query('Price > 50 and Category == "Groceries"') # getting price > 50 and category = groceries
```

```python
d1.iloc[0]
```

```python
d1[d1['Gender'] == "Male"]
```

```python
s = d1.groupby('City').sum()['Total']
s
```

```python
import matplotlib.pyplot as plt
location = d1.groupby('Location')
l =[x for x,i in location]
l
l1 = d1.groupby('Location').sum()['Total']
l1
```

```python
plt.bar(l,l1,color='purple',label='Location')
plt.legend()
plt.title("Total Sales of each location\n")
plt.xlabel("Location")
plt.ylabel("Total sales")
plt.show()
```

```python
colors = ['purple', 'lightblue', 'pink']
plt.pie(l1,labels = l,colors=colors,autopct='%1.1f%%')
plt.title("Percentage of sales in each location")
plt.show()
```

```python
ls= d1.groupby(['Gender','Location']).count()['Invoice ID']
ls
```

```python
u = ls.unstack(level = 0)
u.plot(kind='bar', color=['violet', 'purple'])
plt.ylabel("count of ID's")
plt.title("Male and Female customers in each Locations\n")
```

```python
pay = d1.groupby('Payment')
c =[x for x,i in pay]
c1 = d1.groupby('Payment').count()['Invoice ID']
c1
```

```python
plt.pie(c1,labels = c,autopct='%1.1f%%',colors=['purple','pink','violet'])
plt.title("Payment method percentage")
```
